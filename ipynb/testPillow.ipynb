## イメージ情報の取得

***

```python
from PIL import Image

imFile = Image.open('jpg/{0}.jpg'.format('sample01'))
(w, h) = imFile.size
img = imFile.load()
for m in range(w):
    for n in range(h):
        (r, g, b) = img[m, n]
        img[m, n] = (255 - r, 255 - g, 255 - b)
imFile
```

## 直線の描画

***

```python
from PIL import Image, ImageDraw, ImageColor

img = Image.new('RGB', (300, 300), (255, 255, 255))
draw = ImageDraw.Draw(img)

for n in range(0, 11):
    draw.line([(50+n*20, 50), (50+n*20,255)], fill=(255, 0, 0), width=n)
    draw.line([(50, 50+n*20), (255,50+n*20)], fill=(0, 255, 255), width=n)
img.save('jpg\{0}.jpg'.format('lines'))
img
```

## 四角形の描画

***

```python
from PIL import Image, ImageDraw, ImageColor

img = Image.new('RGB', (300, 300), (255, 255, 255))
draw = ImageDraw.Draw(img)
draw.rectangle([0, 0, 299, 299], outline=(255, 0, 0))
for n in range(10):
    draw.rectangle([50+n*10, 50+n*10,250-n*10,250-n*10], 
                   fill=(255-n*25, 255-n*25, 255-n*25))
img.save('jpg\{0}.jpg'.format('rectangle'))
img
```

## 円の描画

***

```python
from PIL import Image, ImageDraw, ImageColor

img = Image.new('RGB', (300, 300), (255, 255, 255))
draw = ImageDraw.Draw(img)
draw.ellipse([0, 0, 299, 299], outline=(255, 0, 0))
for n in range(10):
    draw.ellipse([50+n*10, 50+n*10, 250-n*10, 250-n*10],
                fill=(n*25, n*25, n*25))
img.save('jpg\{0}.jpg'.format('ellipse'))
img
```

## テキストの描画

***

```python
# from PIL import Image, ImageDraw, ImageColor, ImageFont
from PIL import Image, ImageDraw, ImageFont

img = Image.new('RGB', (300, 300), (255, 255, 255))
draw = ImageDraw.Draw(img)
draw.ellipse([50, 50, 250, 250], fill=(255, 100, 100))
font = ImageFont.truetype('arial.ttf', 40)
draw.text((10, 50), 'Hello!', fill=(0, 0, 255))
draw.text((10, 100), 'This is sample.', font=font, fill=(0, 255, 255))
img.save('jpg\{0}.jpg'.format('text'))
img
```

## 既存ファイルへの描画

***

```python
from PIL import Image, ImageDraw, ImageColor, ImageFont

img = list()
img.append(Image.open('jpg\{0}.jpg'.format('ellipse')))
(w, h) = img[0].size
img.append(img[0].resize((w//3, h//3)))
img[0].paste(img[-1], (0, 0))
img[0].paste(img[-1], (w//3, h//3))
img[0].paste(img[-1], (w//3*2, h//3*2))
img[0].save('jpg\{0}.jpg'.format('multiEllipse'))
img[0]
```

## セピア化

***

```python
from PIL import Image

img = list()
img.append(Image.open('jpg\{0}.jpg'.format('sample01')))
gry = img[0].convert('L')
r = gry.point(lambda x: x * 1.0)
g = gry.point(lambda x: x * 0.75)
b = gry.point(lambda x: x * 0.5)
img.append(Image.merge('RGB', (r,g,b)))
img[-1]
```

## ポスタライズ

***

```python
from PIL import Image, ImageOps

img = list()
img.append(Image.open('jpg\{0}.jpg'.format('sample01')))
img.append(ImageOps.posterize(img[0], 4))
img[-1]
```

```python
from PIL import Image, ImageOps

img = list()
img.append(Image.open('jpg\{0}.jpg'.format('sample01')))
img.append(ImageOps.equalize(img[0]))
img[-1]
```
