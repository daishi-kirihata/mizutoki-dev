## 事前準備（リストデータ）

```python
import numpy

arr = list()
arr.append(['山田'] * 5 + ['田中'] * 5 + ['佐藤'] * 5 + ['鈴木'] * 5 + ['高橋'] * 5 + ['武田'] * 5 + ['中西'] * 5 + ['江口'] * 5 + ['高野'] * 5 + ['小川'] * 5)
arr.append(['国語', '数学', '英語', '理科', '社会'] * 10)
arr.append([numpy.random.randint(0, 100) for i in range(50)])

for data in arr:
    print(data)
```

## DataFrameの作成

***

```python
import pandas

columns = ['名前', '教科', '点数']
data = list(zip(arr[0], arr[1], arr[2]))
dataFrame = pandas.DataFrame(data=data, columns=columns)
dataFrame.info()
dataFrame[:5]
```

## グループ

***

```python
import pandas

columns = ['名前', '教科', '点数']
data = list(zip(arr[0], arr[1], arr[2]))
dataFrame = pandas.DataFrame(data=data, columns=columns)

dataGroup = dataFrame.groupby('教科')
dataGroup.first()
```

## 特定グループのみ取得

***

```python
import pandas

columns = ['名前', '教科', '点数']
data = list(zip(arr[0], arr[1], arr[2]))
dataFrame = pandas.DataFrame(data=data, columns=columns)

dataGroup = dataFrame.groupby('名前')
dataGroup.get_group('武田').sort_values(['点数'], ascending=False)
```

## グループ関数

## グループソート

***

```python
import pandas

columns = ['名前', '教科', '点数']
data = list(zip(arr[0], arr[1], arr[2]))
dataFrame = pandas.DataFrame(data=data, columns=columns)

dataGroup = dataFrame.groupby('名前')
dataGroup.sum().sort_values(['点数'], ascending=False)
```

## 集計

***

```python
import pandas

columns = ['名前', '教科', '点数']
data = list(zip(arr[0], arr[1], arr[2]))
dataFrame = pandas.DataFrame(data=data, columns=columns)

dataGroup = dataFrame.groupby('教科')
dataGroup.agg(['count', 'median', 'max', 'min', 'mean', 'var', 'std'])
```
