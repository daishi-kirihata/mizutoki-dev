## グラフの描画

***

```python
import numpy
from matplotlib import pyplot

graphName = 'x'
x = numpy.array(range(0, 10))
y = x

pyplot.plot(x, y)
pyplot.savefig('jpg/{0}.jpg'.format(graphName))
pyplot.show()
```

## sin曲線の描画

***

```python
import numpy
from matplotlib import pyplot

graphName = 'sin(x)'
x = numpy.arange(-numpy.pi, numpy.pi, 0.1)
y = numpy.sin(x)

pyplot.plot(x, y)
pyplot.savefig('jpg/{0}.jpg'.format(graphName))
pyplot.show()
```

## 複数のグラフの描画

***

```python
import numpy
from matplotlib import pyplot

graphNames = list()
graphNames.append('sin(x)')
graphNames.append('cos(x)')
x = numpy.arange(-2 * numpy.pi, 2 * numpy.pi, 0.1)
plots = list()
plots.append(numpy.sin(x))
plots.append(numpy.cos(x))

for plot in plots:
    pyplot.plot(x, plot)
pyplot.savefig('jpg\{0}.jpg'.format('-'.join(graphNames)))
pyplot.show()
```

## 凡例の追加

***

```python
import numpy
from matplotlib import pyplot

x = numpy.arange(-2 * numpy.pi, 2 * numpy.pi, 0.1)
plots = list()
plots.append((numpy.sin(x), 'sin(x)'))
plots.append((numpy.cos(x), 'cos(x)'))

for plot in plots:
    pyplot.plot(x, plot[0], label=plot[1])

pyplot.legend()
pyplot.show()
```

## グラフ表示の設定

***

```python
import numpy
from matplotlib import pyplot

x = numpy.arange(-2 * numpy.pi, 2 * numpy.pi, 0.1)
plots = list()
plots.append((numpy.sin(x), 'sin(x)'))
plots.append((numpy.cos(x), 'cos(x)'))

for plot in plots:
    pyplot.plot(x, plot[0], label=plot[1])

pyplot.legend()
pyplot.title('sin(x), cos(x)')
pyplot.xlabel('degree')
pyplot.ylabel('value')
pyplot.show()
```

## グリッド表示

***

```python
import numpy
from matplotlib import pyplot

x = numpy.arange(-2 * numpy.pi, 2 * numpy.pi, 0.1)
plots = list()
plots.append((numpy.sin(x), 'y = sin(x)'))
plots.append((numpy.cos(x), 'y = cos(x)'))

for plot in plots:
    pyplot.plot(x, plot[0], label=plot[1])

pyplot.legend()
pyplot.title('sin(x), cos(x)')
pyplot.xlabel('degree')
pyplot.ylabel('value')
pyplot.grid(which='both', axis='x', color='#0000ff', alpha=0.25, linestyle='-', linewidth=1)
pyplot.grid(which='both', axis='y', color='#0000ff', alpha=0.25, linestyle=':', linewidth=1)
pyplot.show()
```

## 描画範囲

***

```python
import numpy
from matplotlib import pyplot

graphNames = list()
graphNames.append('sin(x)')
graphNames.append('cos(x)')

x = numpy.arange(-2 * numpy.pi, 2 * numpy.pi, 0.1)
plots = list()
plots.append((numpy.sin(x), 'y = sin(x)'))
plots.append((numpy.cos(x), 'y = cos(x)'))

for plot in plots:
    pyplot.plot(x, plot[0], label=plot[1])

pyplot.legend()
pyplot.title(', '.join(graphNames))
pyplot.xlabel('degree')
pyplot.ylabel('value')
pyplot.grid(which='both', axis='x', color='#0000ff', alpha=0.25, linestyle='-', linewidth=1)
pyplot.grid(which='both', axis='y', color='#0000ff', alpha=0.25, linestyle=':', linewidth=1)
pyplot.axis([-7, 7, -1.5, 1.5])
pyplot.savefig('jpg/{0}.jpg'.format('-'.join(graphNames)))
pyplot.show()
```

## グラフの形状

***

```python
import numpy
from matplotlib import pyplot

x = numpy.arange(-2 * numpy.pi, 2 * numpy.pi, 0.2)
plots = list()
plots.append((numpy.sin(x), 'm^', 'y = sin(x)'))
plots.append((numpy.cos(x), 'co', 'y = cos(x)'))

for plot in plots:
    pyplot.plot(x, plot[0], plot[1], label=plot[2])

pyplot.legend()
pyplot.title('sin(x), cos(x)')
pyplot.xlabel('degree')
pyplot.ylabel('value')
pyplot.grid(which='both', color='#0000ff')
pyplot.show()
```

## 散布図

***

```python
import numpy
from matplotlib import pyplot
from scipy.stats import norm

graphName = '散布図'
(n, sigma, mu) = (300, 10, 50)
x = numpy.random.randn(n) * sigma + mu
y = numpy.random.randn(n) * sigma + mu
c = numpy.random.rand(n)
s = numpy.random.rand(n)**2 * numpy.pi + 100

pyplot.scatter(x, y, s=s, c=c, alpha=0.25)
pyplot.savefig('jpg\{0}.jpg'.format(graphName))
pyplot.show()
```
